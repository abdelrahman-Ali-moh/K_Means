```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
df=make_blobs(n_samples=250,n_features=2,centers=5,cluster_std=2,random_state=101)
```

```python
df[0].shape
```

```python
plt.scatter(df[0][:,0],df[0][:,1],c=df[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=3)
```

```python
kmeans.fit(df[0])
```

```python
kmeans.cluster_centers_
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,sharey=True,figsize=(10,6))

ax1.set_title('k means')
ax1.scatter(df[0][:,0],df[0][:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title('orginal')
ax2.scatter(df[0][:,0],df[0][:,1],c=df[1],cmap='rainbow')
```
